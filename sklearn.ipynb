```python
from jieba import lcut
from gensim import corpora
import pandas as pd

texts_path = r"C:\Users\user\Desktop\py\衍伸書目\台灣營養健康狀況變遷調查____0629\PROQ-台灣營養健康狀況變遷調查-比對完成.xlsx"
sentence_path = r"C:\Users\user\Desktop\py\衍伸書目\台灣營養健康狀況變遷調查____0629\台灣營養健康狀況變遷調查-訓練集.xlsx"
texts_all = pd.read_excel(texts_path)


# 訓練集

texts = texts_all['Abstract'].astype(str)


labels = texts_all['判斷'].astype(str)


# 分詞
ls_of_words = [lcut(text) for text in texts]

# 構造詞典
dictionary = corpora.Dictionary(ls_of_words)
dt = dictionary.token2id
length = len(dt)

# one hot representation
for word, wid in dt.items():
    vector = [0] * length
    vector[wid] = 1

# 句向量
ls_of_wid = []
for words in ls_of_words:
    vector = [0] * length
    for word in words:
        vector[dt[word]] += 1
    ls_of_wid.append(vector)

# 資料切分
from sklearn.model_selection import train_test_split
train_labels, test_labels, train_wids, test_wids = train_test_split(labels, ls_of_wid)

# 貝葉斯模型訓練
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()  # 樸素貝葉斯分類器
classifier.fit(train_wids, train_labels)

# 模型測評
score = classifier.score(test_wids, test_labels)
print(score)


sentence_all = pd.read_excel(sentence_path)


sentence = sentence_all['Abstract']


sentence_len = len(sentence)

sentence_len_list = range(0, sentence_len)

res = []
for i in sentence_len_list:
    wids = [0] * length
    for wid in dictionary.doc2idx(lcut(str(sentence[i]))):
        wids[wid] += 1
    res.append([classifier.predict([wids])])
    print("第"+str(i+1)+"筆資料分析中")
    #print('句向量', wids)
    #print('預測值', classifier.predict([wids]))
    

```

```python
res_all = pd.DataFrame()
result_df = pd.DataFrame(res)
res_all = res_all.append(result_df)

SaveFile_Path = r'C:\Users\user\Desktop\py'
SaveFile_Name = r'自動判定'
res_all.to_csv(SaveFile_Path+'\\'+ SaveFile_Name + ".csv", encoding = "utf_8_sig", index = False)
```
