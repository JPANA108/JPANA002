```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import pandas as pd
#pd.set_option('display.max_columns', df.shape[0]+1)
#pd.set_option('display.max_rows', None)


def fuzzy_merge(df_1, df_2, key1, key2, threshold=90, limit=2):
    """ param df_1: survey(要比的)
        param df_2: SRDA全2(被比的)
        param key1: 變項名稱設定為 「原始題目」
        param key2: 變項名稱設定為 「原始題目」
        param threshold: 比對精準度 設定為90
        param limit: the amount of matches that will get returned, these are sorted high to low:return: dataframe with boths keys and matches
    """
    s = df_2[key2].tolist()
    m = df_1[key1].apply(lambda x: process.extract(x, s, limit=limit))
    df_1['模糊比對'] = m
    
    m2 = df_1['模糊比對'].apply(lambda x: [i[0] for i in x if i[1] >= threshold][0] if len([i[0] for i in x if i[1] >= threshold]) > 0 else '')
    df_1['模糊比對'] = m2
    return df_1

SRDA_path = r'C:\Users\user\Desktop\py\衍伸書目\台灣營養健康狀況變遷調查____0629\營養all.xlsx'
survey_path = r'C:\Users\user\Desktop\py\衍伸書目\台灣營養健康狀況變遷調查____0629\營養all.xlsx'
SRDA = pd.read_excel(SRDA_path)
survey= pd.read_excel(survey_path)
print('共有'+str(len(SRDA))+'筆要清查')

rs= pd.DataFrame()
for i in range(0, len(SRDA)):
    SRDA = pd.read_excel(SRDA_path)
    survey= pd.read_excel(survey_path)
    survey = survey.iloc[[i],:]
    SRDA = SRDA.drop(index=i)
    print(i)
    df = fuzzy_merge(survey, SRDA, '原始題目', '原始題目', threshold=87)
##################################################################SRDA大清查就設定 87
    rs = pd.concat([rs, df])


#排除檔過濾
del_path = r'C:\Users\user\Desktop\書目勘誤\ALL_SRDA_排除檔.xlsx'
del_SRDA = pd.read_excel(del_path)


finalstepclean = rs.merge(del_SRDA, how='left',on = '原始題目' , indicator=True).loc[lambda x : x['_merge'] == 'left_only']

print(finalstepclean)


SaveFile_Path = r'C:\Users\user\Desktop\書目勘誤'  #要儲存檔案的路徑
SaveFile_Name = r'SRDA清消結果'  #儲存的檔名

finalstepclean.to_excel(SaveFile_Path + '\\' + SaveFile_Name +
                   ".xlsx",
                   encoding="utf_8_sig",
                   index=False)
print('比對成功')


SaveFile_Path = r'C:\Users\user\Desktop\書目勘誤'  #要儲存檔案的路徑
SaveFile_Name = r'ALL_SRDA_排除檔'  #儲存的檔名

rs.to_excel(SaveFile_Path + '\\' + SaveFile_Name +
                   ".xlsx",
                   encoding="utf_8_sig",
                   index=False)
print('比對成功')  
```
