```python
import pandas as pd
A_path = r'C:\Users\user\Desktop\HELP\clean1209-5-累加.xlsx'  
B_path = r'C:\Users\user\Desktop\HELP\臺灣經驗實證調查資料庫使用經驗_寄發名單_2018_20221116 (1).xlsx'  

A = pd.DataFrame(pd.read_excel(A_path))  #GRB名冊
send = pd.DataFrame(pd.read_excel(B_path))  #MOST爬蟲結果

A['Email'] = A['Email'].str.strip()
send['會員帳號'] = send['會員帳號'].str.strip()

merge1 = pd.merge(A,
                  send,
                  left_on='Email',
                  right_on='會員帳號',
                  how='outer')

merge1
```
